# Template for templates

This document is meant to serve as a guide for writing `mdx2.report` templates. The first cell has a title and a short description. 

Templates are executed using [papermill](https://github.com/nteract/papermill). 

## Parameters

Typically, the next cell will contain the input parameters. This cell should have the tag "parameters". Papermill will insert another cell, below this one, with the tag "injected-parameters". 

````python
# These metadata fields are injected by mdx2.report, regardless of the template. Don't change this line.
_metadata: dict = {}

# Below should be the parameters specific to this notebook. Supported types are int, float, str, bool, and tuple.
#
# This example is taken from "scaling_model.ipynb" in the insulin_multi_crystal example.
#
# Type hints and comments are parsed by papermill.inspect_notebook. Pytest uses these to check that the mdx2.reports CLI
# is correctly passing parameters to the notebook. Setting the default value to None signifies a required parameter.

input_files: list[str] = None # list of input file paths: required
model_names: list[str] = ['scaling_model','offset_model','absorption_model','detector_model'] # models to include
pi: float = 3.14159 # numerical value of pi.
````

````python
# pretty-print the metadata and parameters
from pprint import pprint
pprint({
    "metadata": _metadata,
    "parameters": {"input_files": input_files, "model_names": model_names}
})
````

## Results

Helper functions are imported from mdx2.report if needed. Example:

```python
from mdx2.report import load_scaling_models
models = [load_scaling_models(f, model_names=model_names) for f in input_files]
```
