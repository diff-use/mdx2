# Template for templates

This document is meant to serve as a guide for writing `mdx2.report` templates. The first cell has a title and a short description. 

Templates are executed using [papermill](https://github.com/nteract/papermill). 

## Parameters

Typically, the next cell will contain the input parameters. This cell should have the tag "parameters". Papermill will insert another cell, below this one, with the tag "injected-parameters". 

````python
# these metadata fields are populated by mdx2.report, regardless of the template. 
# Typically they are not used by the notebook itself, they are printed to provide context if the notebook is published.
author = None
date_created = None
mdx2_version = None
notebook_template = None
environment = None
working_directory = None

# below should be the parameters specific to this notebook. This example is taken from "scaling_model.ipynb" in the insulin_multi_crystal example

input_files = [] # list of input file paths
model_names = [] # list of scaling model names to include
````

````python
# pretty-print the metadata and parameters
from pprint import pprint
pprint({
    "metadata": {"author": author, 
                 "date_created": date_created, 
                 "mdx2_version": mdx2_version,
                 "notebook_template": notebook_template, 
                 "environment": environment, 
                 "working_directory": working_directory},
    "parameters": {"input_files": input_files, "model_names": model_names}
})
````

## Results

Helper functions are imported from mdx2.report if needed. Example:

```python
from mdx2.report import load_scaling_models
models = [load_scaling_models(f, model_names=model_names) for f in input_files]
```
